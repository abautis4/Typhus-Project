```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

## Datasets

Directory where the files to be used are located and the name of the files.

```python
directory = '/home/abraham/Documents/Python/Notebooks/'
file = 'dataset.csv'
filepath = directory + file
```

Once we have the file path we load the files as dataframes.

```python
df_patients   = pd.read_csv(filepath)
df_patients
```

```python
columns = df_patients.columns.to_list()
```

```python
columns.remove('MRN')
columns.remove('PATIENT HAD TYPHUS?')
columns.remove('PATIENT HAD MISC?')
```

```python
df_analysis = pd.DataFrame(columns=['Feature', 'Male', 'Female', 'Median', 'IQR1', 'IQR3', 'Yes'])
for column in columns:
    feature_typhus = column + ' - Typhus'
    feature_misc = column + ' - MIS-C'
    
    typhus_male = np.nan
    typhus_female = np.nan
    misc_male = np.nan
    misc_female = np.nan
    typhus_yes = np.nan
    misc_yes = np.nan
    typhus_median = np.nan
    typhus_iqr1 = np.nan
    typhus_iqr3 = np.nan
    misc_median = np.nan
    misc_iqr1 = np.nan
    misc_iqr3 = np.nan
    
    if column == 'Gender':
        typhus_male = df_patients['Gender'].loc[(df_patients['PATIENT HAD TYPHUS?']=='Yes')&(df_patients['Gender']=='Male')].value_counts()
        typhus_female = df_patients['Gender'].loc[(df_patients['PATIENT HAD TYPHUS?']=='Yes')&(df_patients['Gender']=='Female')].value_counts()
        misc_male = df_patients['Gender'].loc[(df_patients['PATIENT HAD TYPHUS?']=='No')&(df_patients['Gender']=='Male')].value_counts()
        misc_female = df_patients['Gender'].loc[(df_patients['PATIENT HAD TYPHUS?']=='No')&(df_patients['Gender']=='Female')].value_counts()
    
    if (len(df_patients[column].unique()) == 2) and (column != 'Gender'):
        typhus_yes = df_patients[column].loc[(df_patients['PATIENT HAD TYPHUS?']=='Yes')&(df_patients[column]=='Yes')].value_counts().to_list()
        misc_yes = df_patients[column].loc[(df_patients['PATIENT HAD TYPHUS?']=='No')&(df_patients[column]=='Yes')].value_counts().to_list()
        
        if len(typhus_yes) == 0:
            typhus_yes = [0]
        if len(misc_yes) == 0:
            misc_yes = [0]
        else:
            typhus_yes = typhus_yes[0]
            misc_yes = misc_yes[0]
    
    if len(df_patients[column].unique()) > 2:
        typhus_median = df_patients[column].loc[df_patients['PATIENT HAD TYPHUS?']=='Yes'].median()
        typhus_iqr1 = df_patients[column].loc[df_patients['PATIENT HAD TYPHUS?']=='Yes'].quantile(0.25)
        typhus_iqr3 = df_patients[column].loc[df_patients['PATIENT HAD TYPHUS?']=='Yes'].quantile(0.75)
        misc_median = df_patients[column].loc[df_patients['PATIENT HAD TYPHUS?']=='No'].median()
        misc_iqr1 = df_patients[column].loc[df_patients['PATIENT HAD TYPHUS?']=='No'].quantile(0.25)
        misc_iqr3 = df_patients[column].loc[df_patients['PATIENT HAD TYPHUS?']=='No'].quantile(0.75)
    
    try:
        misc_male = float(misc_male)
        misc_female = float(misc_female)
        misc_yes = misc_yes
        typhus_male = float(typhus_male)
        typhus_female = float(typhus_female)
        typhus_yes = typhus_yes
        
    except:
        pass
    
    dictionaryM = {'Feature': feature_misc,
                   'Male': misc_male,
                   'Female': misc_female,
                   'Median': misc_median,
                   'IQR1': misc_iqr1,
                   'IQR3': misc_iqr3,
                   'Yes': misc_yes}
    
    dictionaryT = {'Feature': feature_typhus,
                  'Male': typhus_male,
                  'Female': typhus_female,
                  'Median': typhus_median,
                  'IQR1': typhus_iqr1,
                  'IQR3': typhus_iqr3,
                  'Yes': typhus_yes}
    
    df_analysis = df_analysis.append(dictionaryM, ignore_index=True)
    df_analysis = df_analysis.append(dictionaryT, ignore_index=True)
```
